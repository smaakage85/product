## Exploratory data analysis
Collect 1,000 observations (pairs of salary ($=y$), age and skills ($=X$))

```python
y, X = simulate_data(n=1000)
```


Plot salary versus age

```python
plt.scatter(x=X['age'], y=y)
plt.xlabel("age (years)")
plt.ylabel("salary (DKK)")
plt.title("Salary versus age")
plt.show()
```

It appears there is a linear relationship between salary and age.

Now plot salary versus skills

```python
plt.scatter(x=X['skills'], y=y)
plt.xlabel("Salary (DKK)")
plt.ylabel("skills")
plt.title("Salary versus skills")
plt.show()
```

It seems that skills and salary are independent.

## Model training
Instantiate and train linear model

```python
m = Model()
m.train(X, y) 
```

Compute predicted salaries

```python
predictions = m.predict(X)
```

Compare predicted and observed salaries

```python
ax = plt.gca()
ax.scatter(X["age"], y, color="b")
ax.scatter(X["age"], predictions, color="r")
plt.xlabel("age")
plt.ylabel("salary (DKK)")
plt.title("Actual versus predicted salary")
plt.gca().legend(("actual", "predicted"))
plt.show()
```

What a beautiful fit!
